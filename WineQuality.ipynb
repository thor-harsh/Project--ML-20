```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("WineQT.csv")

X = data.drop('quality', axis=1)
y = data['quality']


scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

```

```python
pca = PCA(n_components=3)
X_pca_train = pca.fit_transform(X_train)
X_pca_test= pca.transform(X_test)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter=ax.scatter(X_pca_train[:, 0], X_pca_train[:, 1], X_pca_train[:, 2], c=y_train, cmap='viridis', label='Training Data')
ax.scatter(X_pca_test[:, 0], X_pca_test[:, 1], X_pca_test[:, 2], c=y_test, cmap='plasma', label='Testing Data')

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

cbar = plt.colorbar(scatter)
cbar.set_label('Wine Quality')

plt.legend(loc='best')
plt.title('PCA of Wine Quality Dataset')



plt.show()
```
